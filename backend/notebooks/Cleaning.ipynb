```python
def table_health_check(df, name):
    print(f"\n{'='*50}")
    print(f"  {name.upper()} — HEALTH CHECK")
    print(f"{'='*50}")
    print(f"  Shape: {df.shape[0]} rows x {df.shape[1]} columns")
    
    # Missing values
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    if len(missing) == 0:
        print("  ✅ No missing values")
    else:
        print(f"  ⚠️  Missing values found:")
        for col, count in missing.items():
            pct = count / len(df) * 100
            print(f"     {col:<35} {count:>5} missing ({pct:.1f}%)")
    
    # Data types
    print(f"\n  Column data types:")
    for col, dtype in df.dtypes.items():
        print(f"     {col:<35} {str(dtype)}")

# Run health check on all tables
table_health_check(riders, "riders")
table_health_check(mpesa,  "mpesa transactions")
table_health_check(sacco,  "sacco records")
table_health_check(loans,  "loan history")
table_health_check(trips,  "app trip data")
table_health_check(apps,   "credit applications")
```

```python
print("AMOUNT COLUMN — BEFORE CLEANING")
print(f"  Total rows:        {len(mpesa_clean):,}")
print(f"  Negative amounts:  {(mpesa_clean['amount'] < 0).sum()}")
print(f"  Zero amounts:      {(mpesa_clean['amount'] == 0).sum()}")
print(f"  Min amount:        KES {mpesa_clean['amount'].min():.2f}")
print(f"  Max amount:        KES {mpesa_clean['amount'].max():.2f}")
print(f"  Mean amount:       KES {mpesa_clean['amount'].mean():.2f}")
print(f"  Median amount:     KES {mpesa_clean['amount'].median():.2f}")

# Step 1: Make all amounts positive
# Direction column (IN/OUT) already tells us the flow
# so negative sign on amount is redundant and confusing
mpesa_clean["amount"] = mpesa_clean["amount"].abs()

# Step 2: Remove zero-amount transactions
# A zero-amount transaction means nothing happened
rows_before = len(mpesa_clean)
mpesa_clean = mpesa_clean[mpesa_clean["amount"] > 0]
rows_removed = rows_before - len(mpesa_clean)
print(f"\n  Removed {rows_removed} zero-amount rows")

# Step 3: Check for extreme outliers
# A single transaction above KES 100,000 for a boda rider is suspicious
suspicious = mpesa_clean[mpesa_clean["amount"] > 100000]
print(f"\n  Transactions above KES 100,000: {len(suspicious)}")
if len(suspicious) > 0:
    print("  Suspicious transactions:")
    display(suspicious[["rider_id","txn_date","amount",
                          "direction","counterparty_type"]])

print(f"\nAMOUNT COLUMN — AFTER CLEANING")
print(f"  Min:    KES {mpesa_clean['amount'].min():.2f}")
print(f"  Max:    KES {mpesa_clean['amount'].max():.2f}")
print(f"  Mean:   KES {mpesa_clean['amount'].mean():.2f}")
print(f"  Median: KES {mpesa_clean['amount'].median():.2f}")
print(f"\n✅ Amount column cleaned")
```

```python
print("DIRECTION COLUMN CHECK")
print()
print("Unique values found:")
print(mpesa_clean["direction"].value_counts())
print()
```

```python
valid_directions = ["IN", "OUT"]
unexpected = mpesa_clean[
    ~mpesa_clean["direction"].isin(valid_directions)
]
print(f"Unexpected direction values: {len(unexpected)}")
```

```python
null_directions = mpesa_clean["direction"].isnull().sum()
print(f"Null direction values: {null_directions}")

if len(unexpected) == 0 and null_directions == 0:
    print("\n Direction column is clean")
else:
    # Fix unexpected values
    mpesa_clean = mpesa_clean[
        mpesa_clean["direction"].isin(valid_directions)
    ]
    print(f"\n  Removed {len(unexpected)} rows with invalid direction")
```

```python
mpesa_clean["is_inflow"] = (mpesa_clean["direction"] == "IN").astype(int)

print(f"\nInflow transactions:  {mpesa_clean['is_inflow'].sum():,}")
print(f"Outflow transactions: {(mpesa_clean['is_inflow']==0).sum():,}")
```

```python
print("COUNTERPARTY TYPE — VALIDATION")
print()

# These are the only valid types we defined in Step 2
VALID_TYPES = [
    "INCOME_IN",
    "FUEL",
    "SACCO_CONTRIB",
    "FAMILY_SEND",
    "DIGITAL_LOAN",
    "AIRTIME",
    "UNCLASSIFIED"
]

print("Types found in data:")
type_counts = mpesa_clean["counterparty_type"].value_counts()
for txn_type, count in type_counts.items():
    pct = count / len(mpesa_clean) * 100
    is_valid = "✅" if txn_type in VALID_TYPES else "⚠️ UNEXPECTED"
    print(f"  {is_valid} {txn_type:<20} {count:>7,} rows  ({pct:.1f}%)")

# Tag any unknown types as UNCLASSIFIED
mpesa_clean.loc[
    ~mpesa_clean["counterparty_type"].isin(VALID_TYPES),
    "counterparty_type"
] = "UNCLASSIFIED"

print(f"\n✅ Counterparty types validated")
print(f"   Any unknown types tagged as UNCLASSIFIED")
```

```python
print("RUNNING BALANCE — CHECK")
print()
print(f"Min balance:    KES {mpesa_clean['running_balance'].min():.2f}")
print(f"Max balance:    KES {mpesa_clean['running_balance'].max():.2f}")
print(f"Mean balance:   KES {mpesa_clean['running_balance'].mean():.2f}")
print(f"Null balances:  {mpesa_clean['running_balance'].isnull().sum()}")

# Count negative balances — these are Fuliza situations
negative_balances = (mpesa_clean["running_balance"] < 0).sum()
very_negative     = (mpesa_clean["running_balance"] < -5000).sum()

print(f"\nNegative balance rows:       {negative_balances}")
print(f"Very negative (<-5000) rows: {very_negative}")

# Very negative balances suggest data error — cap them
if very_negative > 0:
    print(f"\n⚠️  Capping balances below -5000 to -5000")
    mpesa_clean["running_balance"] = mpesa_clean["running_balance"].clip(lower=-5000)

# Fill any null balances with 0
mpesa_clean["running_balance"] = mpesa_clean["running_balance"].fillna(0)

print(f"\n✅ Running balance cleaned")
print(f"   Min balance after cleaning: KES {mpesa_clean['running_balance'].min():.2f}")
```

```python
print("=" * 55)
print("  M-PESA TABLE — FINAL CLEANED STATE")
print("=" * 55)
print(f"\n  Total rows:        {len(mpesa_clean):,}")
print(f"  Total riders:      {mpesa_clean['rider_id'].nunique():,}")
print(f"  Date range:        {mpesa_clean['txn_date'].min()} "
      f"to {mpesa_clean['txn_date'].max()}")
print(f"  Missing values:    {mpesa_clean.isnull().sum().sum()}")
print()
print("  FINAL COLUMNS:")
for col in mpesa_clean.columns:
    dtype = mpesa_clean[col].dtype
    nulls = mpesa_clean[col].isnull().sum()
    print(f"  ✅  {col:<30} dtype={str(dtype):<10} nulls={nulls}")

print()
print("  TRANSACTION TYPE BREAKDOWN:")
for txn_type, count in mpesa_clean["counterparty_type"].value_counts().items():
    bar = "█" * int(count / 1000)
    print(f"  {txn_type:<20} {count:>7,}  {bar}")

print()

# Quick sanity check on income transactions
income = mpesa_clean[mpesa_clean["counterparty_type"] == "INCOME_IN"]
print(f"  Avg daily income transaction: KES {income['amount'].mean():.0f}")
print(f"  Avg fuel transaction:         "
      f"KES {mpesa_clean[mpesa_clean['counterparty_type']=='FUEL']['amount'].mean():.0f}")

print()
print("✅ M-Pesa table is clean and ready for feature engineering")
print()

# Save the cleaned version
mpesa_clean.to_csv("mpesa_clean.csv", index=False)
print("💾 Saved as mpesa_clean.csv")
```
